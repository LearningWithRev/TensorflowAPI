# Simple Image Cropping with Tensorflow
* tf.image.central_crop()
* tf.image.random_crop()
* tf.image.stateless_random_crop()
* tf.image.resize_with_crop_or_pad()

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
```

### Load Image normally and display it

```python
img = cv2.imread('goose.png', -1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title(f'0 {img.shape}')
plt.imshow(img)
print(img.shape)
print(img.dtype)
```

### Central Crop

```python
c50 = tf.image.central_crop(img, .5)
c50_shape = c50.shape

c20 = tf.image.central_crop(img, .8)
c20_shape = c20.shape

fig = plt.figure(figsize=(10,10))

fig.add_subplot(2, 2, 1)
plt.imshow(c50)
plt.title(f'50 {c50_shape}')

fig.add_subplot(2, 2, 2)
plt.imshow(c20)
plt.title(f'20 {c20_shape}')
```

### Random Crop

```python
c1 = tf.image.random_crop(img, (2000,2000,3))
c2 = tf.image.random_crop(img, (2000,2000,3))

c3 = tf.image.stateless_random_crop(img, (2000,2000,3), seed=[431,2])
c4 = tf.image.stateless_random_crop(img, (2000,2000,3), seed=[431,2])
fig = plt.figure(figsize=(10,10))

fig.add_subplot(2, 2, 1)
plt.imshow(c1)
plt.title(f'Crop 1')

fig.add_subplot(2, 2, 2)
plt.imshow(c2)
plt.title(f'Crop 2')


fig.add_subplot(2, 2, 3)
plt.imshow(c3)
plt.title(f'Stateless 1')

fig.add_subplot(2, 2, 4)
plt.imshow(c4)
plt.title(f'Stateless 2')
```

### Resize Crop or Pad

```python
crop = tf.image.resize_with_crop_or_pad(img, 1500,1500)
pad = tf.image.resize_with_crop_or_pad(img, 4000, 2000)

fig = plt.figure(figsize=(10,10))

fig.add_subplot(2, 2, 1)
plt.imshow(crop)
plt.title(f'Crop')

fig.add_subplot(2, 2, 2)
plt.imshow(pad)
plt.title(f'Padding')
```
